```python
import pandas as pd

# Load the cleaned datasets
math_df = pd.read_csv('./data/student_math_clean.csv')
por_df = pd.read_csv('./data/student_portuguese_clean.csv')
```

```python
# Check for missing values
print("Missing values in math dataset:\n", math_df.isnull().sum())
print("\nMissing values in Portuguese dataset:\n", por_df.isnull().sum())

```

```python
# Merge datasets on student_id (only keep matches from both)
merged_df = pd.merge(math_df, por_df, on='student_id', suffixes=('_math', '_por'))

# Check result
print("Merged dataset shape:", merged_df.shape)
merged_df.head()
```

```python
# Save to CSV for future use
merged_df.to_csv('./data/student_merged.csv', index=False)
```

```python
def classify_performance(avg_grade: float) -> str:
    """
    Classifies performance based on average final grade.

    Parameters:
        avg_grade (float): The student's average final grade.

    Returns:
        str: Performance category ("Excellent", "Good", "Needs Improvement", or "Failing").
    """
    if avg_grade >= 16:
        return "Excellent"
    elif avg_grade >= 13:
        return "Good"
    elif avg_grade >= 10:
        return "Needs Improvement"
    else:
        return "Failing"

# Apply it
merged_df['performance_label'] = merged_df['average_final_grade'].apply(classify_performance)
```

```python
merged_df['average_final_grade'] = (
    merged_df[['final_grade_math', 'final_grade_por']].mean(axis=1)
)
```

```python
# 1. Create average_final_grade column
merged_df['average_final_grade'] = (
    merged_df[['final_grade_math', 'final_grade_por']].mean(axis=1)
)

# 2. Define the classification function
def classify_performance(grade: float) -> str:
    if grade >= 15:
        return "Excellent"
    elif grade >= 10:
        return "Satisfactory"
    else:
        return "Failing"

# 3. Apply the function to create performance_label column
merged_df['performance_label'] = merged_df['average_final_grade'].apply(classify_performance)

# 4. Display the result
print(merged_df[['student_id', 'average_final_grade', 'performance_label']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean style
sns.set_style("whitegrid")

# Bar plot
plt.figure(figsize=(8, 5))
sns.countplot(x='performance_label', data=merged_df, palette='pastel')
plt.title('Number of Students by Performance Category')
plt.xlabel('Performance Category')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='performance_label', hue='performance_label', data=merged_df, palette='pastel', legend=False)
plt.title('Number of Students by Performance Category')
plt.xlabel('Performance Category')
plt.ylabel('Number of Students')
plt.show()

```
